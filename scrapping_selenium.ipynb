```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By

chromedriver_path = "chromedriver.exe"

service = Service(executable_path=chromedriver_path)
driver = webdriver.Chrome(service=service)


```

```python
URL = "https://www.amazon.com/s?k=bicicletas&__mk_es_US=%C3%85M%C3%85%C5%BD%C3%95%C3%91&crid=1LK5BUSB5F206&sprefix=bicicletas%2Caps%2C316&ref=nb_sb_noss_1"  
```

```python
driver.get(URL)  
```

```python
def get_html(url, selector, sleep_time=20):
    driver.get(url)
    time.sleep(sleep_time)
    element = driver.find_element(By.CSS_SELECTOR, selector) #Aprovechamos el find_element para obtener el elemento de css que nos devuelve el html
    return element.get_attribute('outerHTML')
```

```python
html_content = get_html(URL, "div.s-main-slot.s-result-list.s-search-results.sg-row") #Obtenemos el html del elemento que contiene los resultados de la busqueda
```

```python
html_content
```

```python
soup = BeautifulSoup(html_content)
```

```python
soup.find_all("h2")
```

```python
for bike in soup.find_all("h2"):
    print(bike.text.strip())
```

```python
products = []

for bike in soup.find_all("h2", class_="a-size-medium a-spacing-none a-color-base a-text-normal"):
    title = bike.get_text(strip=True)
    
    if title and len(title) > 20:
        products.append(title)

for i, product in enumerate(products, 1):
    print(f"{i}. {product}")
```
